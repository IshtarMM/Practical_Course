```r
# make 3 otus from 9 sample
Otu1 <- c(38.9, 61.2, 73.3, 21.8, 63.4, 64.6, 48.4, 48.8, 48.5)

Otu2 <- c(2.8, 4, 3.4, 4, 9.4, 73.3, 67.3, 61.3, 62.4) 

Otu3 <- c(67.8, 60, 63.4, 76, 89.4, 20.3, 1.3, 2.3, 0.4) 
```

```r

# Create a data frame

my_data <- data.frame( 
  microbes = rep(c("Bacillus", "Fungi1","virus"), each = 9),
   Abundance = c(Otu1,  Otu2,Otu3),
  group = c(rep("control", each = 5),rep("un-infected", each = 5),
            rep("infected", each = 5), 
            rep("control", each = 4),rep("un-infected", each = 4),
            rep("infected", each = 4)),
  body_part = c(rep("skin", each = 13),rep("gut", each = 14)))

my_data 
```

```r
shapiro.test(my_data$Abundance) #Normality Test in R
```

<span style="color:green">From the output, the p-value < 0.05 implying that the distribution of the data are significantly different from normal distribution. In other words, we can assume the not-normality.

```r
library("ggpubr")
ggdensity(my_data$Abundance, 
          main = "Density plot of the OTUs abundance",
          xlab = "OTUs Abundance")


gghistogram(my_data, x = "Abundance",
   add = "mean", rug = TRUE , color = "body_part", fill = "body_part",
   palette = c("#00AFBB", "#E7B800"))
```

```r
#statistic in R
anova_result <- aov(Abundance ~ group , my_data)
```

```r
summary(anova_result)
```

```r
TukeyHSD(anova_result)
```

```r
t.test(Abundance~body_part, data=my_data , exact = FALSE )
```

```r
wilcox.test(Abundance~body_part, data = my_data , exact = FALSE )
```
